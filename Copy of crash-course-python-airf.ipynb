## 2. How do we make the computer decide? → **Conditionals**

Often the computer should choose what to do based on a value.

A conditional makes that choice, using `if`, `elif`, `else`, and comparison operators (`>=`, `>`, `==`, and so on).

````python
score = 40
````

- Cases
  - Greater than or equal to 50
  - Other wise

````python
if score >= 50:
  print("Pass")
else:
  print("Fail")
````

You can test several cases with `elif`:

- Cases
  - Greater than or equal to 90 => distinction
  - Greater than or equal to 50 but less than 90 => pass
  - Otherwise => fail

`elif` => else if

````python
if score >= 90: #
  print("Distinction")
elif score >= 50:
  print("Pass")
elif score >= 30:
  print("Check on student")
else:
  print("Fail")
````

**What is happening:** Python checks each condition top to bottom and runs the first block that is true. The colon : opens the block, and the indented line belongs to it.

`>=` means greater than or equal to; `==` (two equals) asks are these equal?, which is different from a single `=` that stores a value.

**Comparisons** ask a true-or-false question. The common ones:

`==` => comparison operator
`=` => assignment

score = 90

score == 90

````python
print(10 > 5)      # True   -> greater than
print(10 < 5)      # False  -> less than
print(10 >= 10)    # True   -> greater than or equal to
print(10 == 10)    # True   -> equal to (note: two equals signs)
print(10 != 5)     # True   -> not equal to
````

**Logical operators** let you combine several conditions into one. The three you need are `and`, `or`, and `not`:

````python
age = 25
has_id = True

# and -> true only if BOTH sides are true
if age >= 18 and has_id:
    print("Allowed in")

# or -> true if EITHER side is true
if age < 13 or age >= 65:
    print("Gets a discount")

# not -> flips true to false and back
if not has_id:
    print("Please bring your ID")
````

Cases:
- Score is greater than 50 **AND** greater than or equal to 90 => distinction

````python
score = 90

if score > 50 and score >= 90:
  print("Distinction")
````

**Task:** for a number n, print "Fizz" if it divides by 3, "Buzz" if it divides by 5, "FizzBuzz" if it divides by both, and the number itself otherwise.

> You will need one new tool, the **modulo operator** `%`, which gives the remainder after division. If the remainder is 0, the number divides evenly:

Cases:
- "Fizz" => divides by 3
- "Buzz" => divides by 5
- "FizzBuzz" => divides by 3 AND divides by 5
- Otherwise

````python
print(10 % 5)    # 0  -> 10 divides evenly by 5
print(10 % 3)    # 1  -> 10 does not divide evenly by 3
````

````python
n = 20

if n % 5 == 0 and n % 3 == 0:
  print("FizzBuzz")
if n % 5 == 0:
  print("Buzz")
elif n % 3 == 0:
  print("Fizz")
else:
  print(n)
````

## 3. How do we store many values, not just one? → **Data structures**

So far each variable has held a single value. A **data structure** is a container that holds many values together in one place, under one name, with its own rules for how you put things in, take them out, and arrange them.

Different structures suit different jobs, so the skill is picking the right container. Python gives you four common ones.

### 3.1. List
An ordered collection of items, written with square brackets `[ ]`:

````python
cities = ["Lagos", "Accra", "Nairobi", "Cape Town"]

# We can perform different operations on lists
print(cities[0])     # Lagos  -> counting starts at 0
print(cities[2])     # Nairobi
print(len(cities))   # 4      -> how many items
````

**Common list operations:**

````python
cities.append("Kigali")     # add an item to the end
cities.remove("Accra")      # remove an item by value
cities[0] = "Ibadan"        # change an item by position
print("Lagos" in cities)    # True/False -> is it in the list?
print(cities[1:3])          # a slice -> items 1 up to (not including) 3
````

### 3.2. Dictionary

Used to store data as key-value pairs. It is written with curly braces `{ }`. Use it when each item has a label:

````python
scores = {
    "Bola": 70,
    "Angela": 92,
    "Kwame": 50,
    "Justina": 65,
}

print(scores["Justina"]) # we can fetch the value for any key
````

**Common dictionary operations:**

````python
scores["Bola"] = 75   # add or update a key
del scores["Bola"]         # remove a key
print("Angela" in scores)      # True/False -> is the key present?
print(scores.keys())          # all the keys
print(scores.values())        # all the values
````

### 3.3. Tuple

Like a list, but **immutable**, meaning once you create it you cannot change, add, or remove items. Written with `( )`:

````python
days = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
print(days[0])       # Monday  -> read by position, counting from 0
print(days[-1])      # Sunday  -> last item

# days[0] = "Tuesday"   # Uncomment to see the error: tuples cannot be changed
````

**Why immutable?** The days of the week are a fixed set, you would never add, remove, or reorder one. A tuple locks them so nothing in your code can change them by accident, which also signals to anyone reading that these are meant to stay constant. Use a list when the contents will change, a tuple when they should stay put.

**Common tuple operations**: you can only read, not change:

````python
print(len(days))         # 7      -> how many items
print(days[1])           # Tue    -> read by position
print("Sunday" in days)     # True   -> is the value present?
````

### 3.4. Set

A collection with no duplicates and no order, written with `{ }`:

````python
languages = {"Python", "R", "Python"}
print(languages)     # {'Python', 'R'}  -> the repeat is dropped
````

**Common set operations:**

````python
languages.add("Julia")          # add an item (ignored if already present)
languages.remove("R")           # remove an item
print("Python" in languages)    # True/False -> membership, very fast
````

````python
# Union, intersection, difference => the real strength of sets
a = {"Python", "R", "SQL"}
b = {"JavaScript", "Julia", "SQL"}

print(a.union(b))     # union -> everything in either: {'Python', 'R', 'JavaScript', 'SQL', 'Julia'}
print(a.intersection(b))  # intersection -> only what is in both: {'SQL'}
print(a.difference(b))  # difference -> in a but not b: {'R', 'Python'}
````

**What is happening:** `union` combines two sets, `intersection` keeps only what they share, and `difference` keeps what the first has that the second does not. These make sets ideal for questions like "who took both courses?" or "what is unique to this group?".

### Strings are sequences too

A string is an ordered sequence of characters, so everything you know about indexing, slicing, and looping from lists works on text as well.

This matters especially for this programme: the courses build towards large language models (LLMs). That is all string work, so the more comfortable you are with strings now, the smoother the rest of the curriculum will feel.

**Common string operations:**

Indexing and slicing:

````python
name = "Ada Lovelace"
print(len(name))     # 12   -> number of characters (spaces count)
print(name[0])       # A    -> first character
print(name[-1])      # e    -> last character
print(name[0:3])     # Ada  -> a slice, characters 0 up to (not including) 3
print(name[4:])      # Lovelace  -> from position 4 to the end
````

Changing case (lowercasing is a very common first step in NLP, so "Word" and "word" count as the same):

````python
text = "Natural Language Processing"
print(text.upper())      # NATURAL LANGUAGE PROCESSING
print(text.lower())      # natural language processing
print(text.title())      # Natural Language Processing  -> first letter of each word
````

Cleaning:

````python
messy = "  hello  "
print(messy.strip())                        # "hello"  -> trims spaces from both ends
print("that's amazing!".replace("!", ""))   # that's amazing  -> remove punctuation
````

Checking what a string contains:

````python
print("data" in "database")          # True   -> is this text inside that text?
print("report.pdf".endswith(".pdf")) # True
print("hello".count("l"))            # 2      -> how many times it appears
print("the cat sat".find("cat"))     # 4      -> position where "cat" starts
````

Splitting and joining:

````python
sentence = "learning python is fun"
words = sentence.split()     # ['learning', 'python', 'is', 'fun'] -> split on spaces
print(len(words))            # 4  -> now you can count the words

joined = " ".join(words)     # 'learning python is fun' -> glue a list back into a string

print(joined)
````

> Splitting a sentence into a list of words is called **tokenisation**, one of the very first steps in NLP, and you will meet it again properly later in the programme.

**What is happening:** most string tools are written as `text.something(...)`, a **method**, which is just a function that belongs to that value. Because a string is a sequence, indexing, slicing, and looping all work on text just as they do on lists.

## 4. How do we repeat a step for every item? → **Loops**

When you want to do the same thing many times, you do not write it out by hand. A **loop** repeats a step across every item in a collection.

Loop over a list:

````python
cities = ["Lagos", "Accra", "Nairobi", "Cape Town"]

for city in cities:
    print(city)
````

Loop over a dictionary:

````python
capitals = {"Nigeria": "Abuja", "Ghana": "Accra", "Kenya": "Nairobi"}

for country in capitals:
    print(country, "->", capitals[country])
````

Loop over a sequence of numbers within a range with `range()`:

````python
for i in range(5):
    print(i)          # 0, 1, 2, 3, 4  -> starts at 0, stops before 5
````

Loop over the characters of a string:

````python
for letter in "AI":
    print(letter)
````

A **while** loop repeats as long as a condition stays true:

````python
count = 3

while count > 0:
    print(count)      # 3, 2, 1
    count = count - 1
````

**What is happening:** `for item in collection` takes each item one at a time, gives it a name, and runs the indented block for each, stopping automatically when it runs out.

A `while` loop keeps going until its condition becomes false, so you must change something inside it, or it never stops.

## 5. How do we reuse our logic without repeating it? → **Functions**

When you find yourself writing the same steps again and again, wrap them up. A function packages steps under a name so you write them once and reuse them anywhere.

````python
def greet(name):
    return "Hello, " + name + "!"

print(greet("Ada"))       # Hello, Ada!
print(greet("Samuel"))    # Hello, Samuel!
````

A function can take more than one input and give a result back:

````python
def add(a, b):
    return a + b

print(add(3, 5))     # 8
````

**What is happening:** `def` names the **function**, the inputs go in the brackets (called **parameters**), and `return` hands a result back to whoever called it. Define it once, then call it as many times as you like with different inputs. Without `return`, a function does its work but gives nothing back.

## 6. How do we avoid reinventing common tasks? → **Libraries**

Many everyday tasks are already solved by others. A library is a collection of ready-made functions you bring in with `import`, so you reuse their work instead of writing it yourself.

````python
import math
print(math.sqrt(144))    # 12.0  -> square root
print(math.pi)           # 3.141592653589793
````

````python
import statistics
readings = [21, 25, 19, 30, 22]
print(statistics.mean(readings))     # 23.4  -> the average
````

You can import just one piece/method from a library:

````python
from random import choice

options = ["Lagos", "Nairobi", "Kigali", "Cape Town", "Morocco"]
print(choice(options))   # picks one random item from a list
````

**What is happening:** `import math` loads the whole library, and you reach inside it with `math.something(...)`. `from random import choice` pulls out just the one function so you can use it by name.

You did not write `sqrt`, `mean`, or `choice`; someone else did, and you reuse them. Knowing a library exists and how to look up what it offers matters more than memorising it, and this is exactly how the AI courses will lean on libraries built for data and machine learning.

## 7. How do we put it all together? → **Thinking in code**

Writing code starts with understanding the problem and thinking, not typing.

Before writing any code, break it down into small steps, and tackle each step on its own. Try to write it in plain language first, then turn each step into a line of code.

> **Task:** Go through a week of temperature readings. Print the average temperature and how many hot days there were.

> **Note:** **Hot** is 30 and above, **Warm** is 20 to 29, **Cool** is below 20.

Before breaking anything down, get clear on two things: what goes **in**, and what should come **out**. Everything in between is just the logic.

**Pseudocode:**

```text
INPUT: readings (a week of temperatures)
OUTPUT: the average temperature, and the number of hot days

SET hot_days to 0

FOR each temp in readings:
    IF temp is 30 or more:
        ADD 1 to hot_days

COMPUTE the average of readings
PRINT the average
PRINT hot_days
```

## 8. What do we do when it breaks? → **Debugging**

You will make errors and run into **bugs**, and that is completely normal, even for experts.

An **error** is not failure; it is the computer telling you where it got confused. The skill is reading the message calmly instead of fearing the red text.

````python
name = "Bola"
age = 70

print(name + age)
````

````python
readings = [21, 25, 19, 30, 22]

total = 0
for temp in readings:
    total = temp          # bug: this replaces total each time instead of adding

average = total / len(readings)
print("Average:", average)     # prints 4.4, which is clearly wrong
````

There is no red error here, that is what makes it interesting. Python runs it happily and prints an average of 4.4 for readings that are all in the 20s, **which cannot be right**.
